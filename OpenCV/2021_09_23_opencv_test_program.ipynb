```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
src = cv2.imread('images/lenna.bmp', cv2.IMREAD_GRAYSCALE)
hist = cv2.calcHist([src], [0], None, [256], [0, 256])
cv2.imshow('src', src)
if cv2.waitKey(0) == 27:
    cv2.destroyAllWindows()
plt.plot(hist)
plt.show()
```
